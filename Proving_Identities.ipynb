```python
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted
from matplotlib_venn import venn3, venn3_circles
from matplotlib import pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
def drawVenn2(A,B,labels = ('A', 'B')):
    '''
    listing elements in Venn diagram of two sets
    '''
    v2=venn2([A,B],set_labels = labels)
    v2.get_label_by_id('01').set_text(', '.join(map(str,B-A)))
    v2.get_label_by_id('10').set_text(', '.join(map(str,A-B)))
```

# Providing Identities

#### $x \in A \land x \in B$ 

1. $A \setminus B = A \cap B^{c}$

- $ x \in A \setminus x \in B$ (the elements in set A that are not in set B)
- $ x \in A \cap B^{c} $ (Union of elements in set A and elements that are not in set B )

```python
A = {'emu otori', 'hoshino\n takanashi'}
B = {'emu otori','yui\n hirasawa'} 

drawVenn2(A,B,labels=('A', 'B'))

# A \ B 
difference = A.difference(B)

# Bᶜ
complement_B = set()
for element in A:
    if element not in B:
        complement_B.add(element)

# A ∩ Bᶜ
intersection = A.intersection(complement_B)

# difference - intersection
is_correct = difference == intersection
print("Is A \setminus B = A \cap B^{c} correct?", is_correct)
```


1. True, since basically anything that is in set A and not in set B is the same as the union of elements in set A and elements that are not in set B

2. B = (B $\cap$ A) $\cup$ (B $\cap$ $A^{c}$)

- $ (x \in B) \cap (x \in A)$ = the similar elements found in set B and elements in set A
- $ (x \in B) \cap A^{c} $  = the similar elements in set B and elements that are not in set A
- $ ((x \in B) \cap (x \in A)) \cup ((x \in B) \cap A^{c}) $ = the union of elements from (similar elements from set A and B) with elements from (similar elements from set B with elements not found in set A)

```python
A = {'emu otori', 'hoshino\n takanashi'}
B = {'emu otori','yui\n hirasawa'} 

drawVenn2(A,B,labels=('A', 'B'))

# (B ∩ A)
intersection_1 = B.intersection(A)

# Aᶜ
complement_A = set()
for element in B:
    if element not in A:
        complement_A.add(element)

# (B ∩ Aᶜ)
intersection_2 = B.intersection(complement_A)

# (B ∩ A) ∪ (B ∩ Aᶜ)
result = intersection_1.union(intersection_2)

# Comparing elements in B with the result
is_correct = result == B
print("Is B = (B ∩ A) U (B ∩ Aᶜ) correct?", is_correct)
```

2. True, since basically if you add all elements from (similar elements of sets A and B) plus (similar elements from B and elements not in set A) you get all exactly all elements from set B

3. B\A = B $\cap$ $A^{c}$

- $ x \in B \setminus x \in A $ = all elements that are in set B not found in set A
- $ x \in B \cap A^{c} $ = similar elements found in set B and elements not from set A

```python
A = {'emu otori', 'hoshino\n takanashi'}
B = {'emu otori', 'yui\n hirasawa'}

drawVenn2(A,B,labels=('A', 'B'))

# B\A
B_minus_A = B - A

# Aᶜ
complement_A = set()
for element in B:
    if element not in A:
        complement_A.add(element)
        
# (B ∩ Aᶜ)
intersection = B.intersection(complement_A)

# Compare the sets
is_correct = B_minus_A == intersection
print("The equation B\\A = B ∩ Aᶜ is ", is_correct)
```

3. True, since basically (elements in B that are not in A) is the same as (the similar elements of set B and elements nor found in set A)

4. A $\cup$ B = A $\cup$ (B $\cap$ $A^{c}$)

- $(B \cap A^{c})$: $ x \in B \cap A^{c} $ = similar elements in set B and elements not from in set A
- $(A \cup (B \cap A^{c}))$: = (the intersection of all elements fron set B with elements not found in set A) in union with set A
- $(A \cup B)$: $ x \in A \cup x \in B $ = the union of all elements fron sets A and B
- $(A \cup B = A \cup (B \cap A^{c}))$: the union of sets A & B, is similar to the union of set A with (the intersection of elements in set B with elements not found in set A)

```python
A = {'emu otori', 'hoshino\n takanashi'}
B = {'emu otori', 'yui\n hirasawa'}

drawVenn2(A,B,labels=('A', 'B'))

# A ∪ B
union_A_B = A.union(B)

# Aᶜ
complement_A = set()
for element in B:
    if element not in A:
        complement_A.add(element)

# B ∩ Aᶜ
intersection_B_A_complement = B.intersection(complement_A)

# A ∪ (B ∩ Aᶜ)
union_A_intersection_B_A_complement = A.union(intersection_B_A_complement)

# checking if both is similar
is_correct = union_A_B == union_A_intersection_B_A_complement
print("The equation A U B = A U (B ∩ Aᶜ) is ", is_correct)
```

4. True, since basically the union of sets A & B is similar to the union of elements in set A with (similar elements in set B with elements not found in set A)

6. |A $\cup$ B| = |A| + |B| - |A $\cap$ B|

- $ |A| \cap |B| $ : the cardinality of the similar elements found in sets A & B
- $ |A| + |B| : cardinality of combined elements of sets A & B
- $ |A| + |B| - |A \cap B| $ : difference between (the cardinality of sum A + B) minus (the cardinality of similar elements found in sets A & B)
- $ |A| \cup |B| $ : all elements found in set A  and set B

```python
A = {'emu otori', 'hoshino\n takanashi'}
B = {'emu otori', 'yui\n hirasawa'}

drawVenn2(A,B,labels=('A', 'B'))

union_cardinality = len(A.union(B)) # 4
A_cardinality = len(A) # 2
B_cardinality = len(B) # 2
intersection_cardinality = len(A.intersection(B)) # 2

is_correct = (union_cardinality == A_cardinality + B_cardinality - intersection_cardinality) # 4 == (2 + 2) - 2
print("The equation is", is_correct)
```

6. True, since basically if you subtract it to the added cardinality of set A and set B minus the intersection of set A and B, it has the same value as all elements of a set A and B
